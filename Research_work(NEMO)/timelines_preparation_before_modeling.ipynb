```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 15, 4
```

```python
df_nemo=pd.read_csv('./formed_timelines/2002_spitz_nemo.csv', sep=';')
df_nemo['dates'] =  pd.to_datetime(df_nemo['dates'], format='%Y-%m-%d')
df_nemo['75.532663_17.902096_v']=df_nemo['75.532663_17.902096']-np.mean(df_nemo['75.532663_17.902096'])
df_nemo['76.006222_24.893752_v']=df_nemo['76.006222_24.893752']-np.mean(df_nemo['76.006222_24.893752'])
df_nemo['75.210923_11.252254_v']=df_nemo['75.210923_11.252254']-np.mean(df_nemo['75.210923_11.252254'])
df_nemo
```

```python
df_rean=pd.read_csv('./formed_timelines/2002_spitz_reanalysis.csv', sep=';')
df_rean['dates'] =  pd.to_datetime(df_rean['dates'], format='%Y-%m-%d')
df_rean['75.532663_17.902096_v']=df_rean['75.532663_17.902096']-np.mean(df_rean['75.532663_17.902096'])
df_rean['76.006222_24.893752_v']=df_rean['76.006222_24.893752']-np.mean(df_rean['76.006222_24.893752'])
df_rean['75.210923_11.252254_v']=df_rean['75.210923_11.252254']-np.mean(df_rean['75.210923_11.252254'])
df_rean
```

```python
plt.plot(df_rean['dates'], df_rean['75.532663_17.902096_v'], c='r', linestyle=':')
plt.plot(df_nemo['dates'], df_nemo['75.532663_17.902096_v'], label='75.532663_17.902096', c='r', linestyle='-')
plt.legend()
plt.grid()
plt.show()
plt.plot(df_rean['dates'], df_rean['76.006222_24.893752_v'], c='g', linestyle=':')
plt.plot(df_nemo['dates'], df_nemo['76.006222_24.893752_v'], label='76.006222_24.893752', c='g', linestyle='-')
plt.legend()
plt.grid()
plt.show()
plt.plot(df_rean['dates'], df_rean['75.210923_11.252254_v'], c='b', linestyle=':')
plt.plot(df_nemo['dates'], df_nemo['75.210923_11.252254_v'], label='75.210923_11.252254', c='b', linestyle='-')
plt.legend()
plt.grid()
plt.show()
```
