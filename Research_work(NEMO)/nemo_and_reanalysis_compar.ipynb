```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
rcParams['figure.figsize'] = 15, 4
```

```python
df_nemo=pd.read_csv('./2002_spitz_nemo.csv', sep=';')
df_nemo['dates'] =  pd.to_datetime(df_nemo['dates'], format='%Y-%m-%d')
df_nemo
```

```python
df_rean=pd.read_csv('./2002_spitz_reanalysis.csv', sep=';')
df_rean['dates'] =  pd.to_datetime(df_rean['dates'], format='%Y-%m-%d')
df_rean['77.325956_29.460769'][df_rean['77.325956_29.460769']==-2147483647]=np.nan
df_rean
```

## Графики для первой координаты

```python
plt.plot(df_rean['dates'], df_rean['75.295924_18.454331'], label='reanalysis')
plt.plot(df_nemo['dates'], df_nemo['75.295924_18.454331'], label='nemo')
plt.title('Lat:75.295924, Lon:18.454331')
plt.legend()
plt.grid()
```

```python
plt.plot(df_rean['dates'], df_rean['75.295924_18.454331']-df_nemo['75.295924_18.454331'])
```

```python
plt.plot(df_rean['dates'], df_rean['75.295924_18.454331'], label='reanalysis')
plt.plot(df_nemo['dates'], df_nemo['75.295924_18.454331']+0.55, label='nemo')
plt.title('Lat:75.295924, Lon:18.454331\nЕсли считать 0.55 систематическим занижением модели NEMO')
plt.legend()
plt.grid()
```
