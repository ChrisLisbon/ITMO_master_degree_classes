### 1. Исследование данных. Целевая переменная - RainTomorrow

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('./weatherAUS.csv')
data['Date'] =  pd.to_datetime(data['Date'], format='%Y-%m-%d')
data=data.sort_values(by='Date')
data
```

```python
data.describe()
```

```python
data.info()
```

## Разделяем категориальные и некатегориальные признаки

```python
qual_params=['WindGustDir', 'WindDir9am', 'WindDir3pm', 'RainToday', 'RainTomorrow']
count_params=['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine', 'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm',
                'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm']
for column in qual_params:
    print('\n'+column)
    print(data[column].unique())
```

## Для категориальных признаков заполняем пропуски наиболее часто встречаемым значением (модой)

```python
for i in qual_params:
    data[i].fillna(data[i].mode()[0], inplace=True)
```

## Для некатегориальных признаков заполняем пропуски средним значением (медианой)

```python
for i in count_params:
    data[i].fillna(data[i].mean(), inplace = True)
```

```python
# смотрим, что пропусков нет
data.isnull().sum()
```

## Создаем новые признаки

```python
data["Date"] = pd.to_datetime(data['Date'])
```

```python
data["year"] = data["Date"].dt.year
data["day"] = data["Date"].dt.day
data["month"] = data["Date"].dt.month
```

## Удаляем неинформативные

```python
data.drop(["Date", "Location"],axis = 1, inplace=True)
```

## Производим бинарную кодировку категориальных признаков

```python
dummies = pd.get_dummies(data[qual_params], drop_first=True)
new_df = pd.concat([data, dummies], axis=1)
new_df.drop(qual_params, axis = 1, inplace = True)
```

```python
new_df
```

## Посмотрим на матрицу корреляций, чтобы узнать какие переменные сильнее всего связаны с целевой

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr_matrix = new_df.corr()
fig, ax = plt.subplots(figsize=(30,20))
ax = sns.heatmap(corr_matrix,
                 annot=True,
                 linewidth=0.5,
                 fmt='.2f',
                 cmap='YlGnBu')
```

### Сильнее всего с целевой переменной связаны: 
- наличие дождя сегодня;
- облачность в 3 и в 9 часов;
- влажность в 3 и в 9 часов;
- давление в 3 и в 9 часов;
- количество осадков;
- солнечность.

```python
predictors=['RainToday_Yes', 'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 
            'Rainfall', 'Sunshine', 'RainTomorrow_Yes']
filtered_df=new_df[predictors]
filtered_df
```

## Производим масштабирование параметров

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

filtered_df=new_df[predictors]
scaling_variables=['Humidity9am','Humidity3pm','Pressure9am','Pressure3pm','Cloud9am','Cloud3pm','Rainfall','Sunshine']

for var in scaling_variables:  
    sc=MinMaxScaler()
    variable=np.array(filtered_df[[var]])
    sc.fit(variable)
    filtered_df[var]=pd.Series(sc.transform(variable).T[0])  
filtered_df
```

## Переводим датасет к формату

```python
from sklearn.model_selection import train_test_split

X = filtered_df.iloc[:, : -1]
y = filtered_df.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

## Производим поиск оптимальных параметров классификатора и степени полиномов по сетке

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
from sklearn.model_selection import StratifiedKFold, GridSearchCV
```

```python
# Логистическая регрессия

parameters_grid = {
       'linear_model__C' : np.arange(1, 3, 0.1), # параметр регуляризации
       'preprocessing__degree': np.arange(1, 3),
   }

classifier = Pipeline([('preprocessing', PolynomialFeatures()), ('linear_model', LogisticRegression())])
cv = StratifiedKFold(n_splits=10, random_state=0)
grid_cv = GridSearchCV(classifier, parameters_grid, scoring = 'accuracy', cv = cv)

grid_cv.fit(X_test, y_test)
print(grid_cv.best_params_)
print(grid_cv.best_score_)
```

```python
# AdaBoost 

parameters_grid = {    
       'preprocessing__degree': np.arange(1, 3),
       'ensemble__n_estimators' : np.arange(99, 101),
   }

classifier = Pipeline([('preprocessing', PolynomialFeatures()), ('ensemble', AdaBoostClassifier())])
cv = StratifiedKFold(n_splits=10, random_state=0)
grid_cv = GridSearchCV(classifier, parameters_grid, scoring = 'accuracy', cv = cv)

grid_cv.fit(X_test, y_test)
print(grid_cv.best_params_)
print(grid_cv.best_score_)
```

```python
# Градиентный бустинг

parameters_grid = {    
       'preprocessing__degree': np.arange(1, 3),
       'ensemble__n_estimators' : np.arange(95, 105),
       'ensemble__learning_rate' : np.arange(0, 1, 0.1),
       'ensemble__max_depth' : np.arange(1, 10),
   }

classifier = Pipeline([('preprocessing', PolynomialFeatures()), ('ensemble', GradientBoostingClassifier())])
cv = StratifiedKFold(n_splits=10, random_state=0)
grid_cv = GridSearchCV(classifier, parameters_grid, scoring = 'accuracy', cv = cv)

grid_cv.fit(X_test, y_test)
print(grid_cv.best_params_)
print(grid_cv.best_score_)
```
